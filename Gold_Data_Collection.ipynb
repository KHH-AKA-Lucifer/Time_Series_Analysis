### History method quick documentation 

**Arguments**

***start*** 
Start date.
default : None 
If specified, override the period argument. 

***end***
End date.
default : current date 

***interval***
Time interval.
default : 1m (1 minute)
Available intervals : 1m, 2m, 5m, 15m, 30m, 60m, 90m, 1h, 1d, 5d, 1wk, 1mo, 3mo

***period***
Time period.
default : 1d (1day)
Available periods : 1d, 5d, 1mo, 3mo, 6mo, 1y, 2y, 5y, 10y, ytd, max

```python
# retrieving one day data of 1 min time interval
# intialize start and end date
start_date = "2024-06-04"
# get historical market data 
gold_one_day = gd.history(start = start_date, interval="1m") 
```
